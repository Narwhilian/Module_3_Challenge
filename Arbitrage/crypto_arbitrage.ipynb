### Step 1: For the bitstamp DataFrame, replace or drop all `NaN`, or missing, values in the DataFrame.

```python
# For the bitstamp DataFrame, replace or drop all NaNs or missing values in the DataFrame
bitstamp_df = bitstamp.dropna().copy(deep=True)

bitstamp_df.isnull().sum()
```

### Step 2: Use the `str.replace` function to remove the dollar signs ($) from the values in the Close column.

```python
# Use the str.replace function to remove the dollar sign, $
new_close = bitstamp_df["Close"].str.replace("$","")

bitstamp_df["Close"] = new_close

bitstamp_df.head()
```

### Step 3: Convert the data type of the Close column to a `float`.

```python
# Convert the Close data type to a float
new_close_type = bitstamp_df.loc[:,"Close"].astype("float")

bitstamp_df["Close"] = new_close_type

bitstamp_df.dtypes
```

### Step 4: Review the data for duplicated values, and drop them if necessary.

```python
# Review the data for duplicate values, and drop them if necessary
bitstamp_df.duplicated().sum()
```

### Step 5: Repeat Steps 1–4 for the coinbase DataFrame.

```python
# Repeat Steps 1–4 for the coinbase DataFrame
# Drops the NaN rows from the coinbase DataFrame
coinbase_df = coinbase.dropna().copy(deep=True)

# Replaces the $ value in the close column 
new_coinbase = coinbase_df["Close"].str.replace("$","")
coinbase_df["Close"] = new_coinbase

# Converts the data type in the close column to 'float64'
new_coinbase_type = coinbase_df.loc[:,"Close"].astype("float")
coinbase_df["Close"] = new_coinbase_type
coinbase_df.dtypes

# Checks for duplicate data in the coinbase DataFrame, showing that no duplicats exist so there is nothing to remove
coinbase_df.duplicated().sum()

```

#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one.
```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one

# creates the filters for each data frame to find positive instances of arbitrage 
arbitrage_spread_early_filter = arbitrage_spread_early > 0
arbitrage_spread_mid_filter = arbitrage_spread_middle > 0
arbitrage_spread_late_filter = arbitrage_spread_late > 0 



# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
# Displays summary statistics for the early dataframe
arbitrage_spread_early[arbitrage_spread_early_filter].describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

# Creates a dataframe of arbitrage spread return for each of the three time frames
spread_return_early= arbitrage_spread_early[arbitrage_spread_early_filter] / bitstamp_sliced.loc["2018-01-16"]
spread_return_middle= arbitrage_spread_middle[arbitrage_spread_mid_filter] / bitstamp_sliced.loc["2018-02-24"]
spread_return_late= arbitrage_spread_late[arbitrage_spread_late_filter] / bitstamp_sliced.loc["2018-03-26"]

# Reviews the spread return for the early DataFrame
spread_return_early.head()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs

# Creates a filter for each time period to only include trades with a profit of greater than 1%
early_profit_filter = spread_return_early >= 0.01
middle_profit_filter = spread_return_middle >= 0.01
late_profit_filter = spread_return_late >= 0.01

# Applies the filter for each time period to create a dataframe that only includes trades with a profit greater than 1%
profitable_trades_early = spread_return_early[early_profit_filter]
profitable_trades_middle = spread_return_middle[middle_profit_filter]
profitable_trades_late = spread_return_late[late_profit_filter]

# Review the first five profitable early trades
profitable_trades_early.head()
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the date early in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp_sliced
profit_middle = profitable_trades_middle * bitstamp_sliced
profit_late = profitable_trades_late * bitstamp_sliced


# Drop any missing values from the profit DataFrames
profit_per_trade_early = profit_early.dropna().copy(deep=True)
profit_per_trade_middle = profit_middle.dropna().copy(deep=True)
profit_per_trade_late = profit_late.dropna().copy(deep=True)

# View the early profit DataFrame
profit_per_trade_early.head()
```
